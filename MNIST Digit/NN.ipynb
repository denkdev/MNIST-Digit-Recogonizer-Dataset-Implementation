```python
#Initialize
Y = Y_train.to_numpy()
X = X_train.to_numpy()

Y_t = Y_test.to_numpy()
X_t = X_test.to_numpy()

X = X.T
Y = Y.T

X_t = X_t.T
Y_t = Y_t.T
```

```python
def calc_Z(W,A,B):
    return W@A + B

def ReLu(Z):
    return np.maximum(0,Z)

def softmax(Z):
    # Subtract max for numerical stability along columns
    Z_exp = np.exp(Z - np.max(Z, axis=0, keepdims=True))
    Z_softmax = Z_exp / np.sum(Z_exp, axis=0, keepdims=True)
    return Z_softmax

def forward_propagation(A_0,W_1,B_1,W_2,B_2):
    Z_1 = calc_Z(W_1,A_0,B_1)
    A_1 = ReLu(Z_1)
    Z_2 = calc_Z(W_2,A_1,B_2)
    A_2 = softmax(Z_2)
    return Z_1, A_1, Z_2 , A_2

def backward_propagation(A_0,W_1,B_1,W_2,B_2,Z_1,A_1,Z_2,A_2,Y):
    dZ_2 = A_2 - Y
    dW_2 = 1/42000 * dZ_2 @ (A_1.T)
    dB_2 = np.mean(dZ_2, axis=1, keepdims=True)
    dB_2 = np.repeat(dB_2,42000, axis = 1)

    dZ_1 = (W_2.T) @ dZ_2 *(Z_1>0)
    dW_1 = 1/42000 * (dZ_1 @ X.T)
    dB_1 = np.mean(dZ_1, axis=1, keepdims=True)
    dB_1 = np.repeat(dB_1,42000, axis = 1)



    return dW_1, dB_1, dW_2, dB_2

def adjust(W_1,B_1,W_2,B_2,dW_1, dB_1, dW_2, dB_2, learning_rate = 0.01):
    W_1 = W_1 - learning_rate * dW_1
    B_1 = B_1 - learning_rate * dB_1
    W_2 = W_2 - learning_rate * dW_2
    B_2 = B_2 - learning_rate * dB_2

    return W_1,B_1,W_2,B_2

def prediction(A_2):
    # Returns the predicted class for each sample (column-wise)
    return np.argmax(A_2, axis=0)

def accuracy(A_2, Y):
    A_out = prediction(A_2)  # Get predictions
    correct = (A_out == Y)    # Compare predicted vs actual labels
    return np.mean(correct)   # Mean of correct predictions


def train(A_0,W_1,B_1,W_2,B_2,Y,learning_rate):
    Z_1,A_1,Z_2,A_2 = forward_propagation(A_0,W_1,B_1,W_2,B_2)
    dW_1, dB_1, dW_2, dB_2 = backward_propagation(A_0,W_1,B_1,W_2,B_2,Z_1,A_1,Z_2,A_2,Y)
    W_1,B_1,W_2,B_2 = adjust(W_1,B_1,W_2,B_2,dW_1, dB_1, dW_2, dB_2, learning_rate)
    print(f'The training Accuracy of the Network is : {accuracy(A_2,Y)}')
    #print(f'The testing Accuracy of the Network is : {}')
    return W_1,Z_1,W_2,Z_2


```

```python
A_0 = X
Z_1 = np.random.rand(10,42000)
W_1 = np.random.rand(10, 784)
B_1 = np.random.rand(10 ,42000)
Z_2 = np.random.rand(10,42000)
W_2 = np.random.rand(10, 10)
B_2 = np.random.rand(10 ,42000)


```

```python
learning_rate = 0.1
num_epochs = 5
for epoch in range(num_epochs):
    W_1,Z_1,W_2,Z_2 = train(A_0,W_1,B_1,W_2,B_2,Y,learning_rate)

```
